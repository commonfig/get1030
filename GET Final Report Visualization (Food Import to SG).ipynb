Make to create 2 text files in a folder (I used "dataNew" folder) to store info for networkprocessing.py (2000.txt and 2020.txt)

```python
x = 0
with open('dataNew/2000.txt', 'w') as f:
    for index, row in df.iterrows():
        while x < int(row.loc['Estimate Ratio']):
            f.write(row.loc['To'] + "," +row.loc['From']+"\n")
            x += 1
        x = 0
```

```python
x = 0
with open('dataNew/2020.txt', 'w') as f:
    for index, row in df2.iterrows():
        while x < int(row.loc['Estimate Ratio']):
            f.write(row.loc['To'] + "," +row.loc['From']+"\n")
            x += 1
        x = 0
```
